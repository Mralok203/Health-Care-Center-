```python
#train test  split 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=20)
```

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
                                
from sklearn.metrics import accuracy_score,confusion_matrix
import numpy as np 

#create adict to store model
models= {
    "SVC":SVC(kernel="linear"),
    "RandomForest": RandomForestClassifier(n_estimators=100,random_state=42),
    "KNeighborsClassifier":KNeighborsClassifier(n_neighbors=5),
    "GradientBoostingClassifier":GradientBoostingClassifier(n_estimators=100,random_state=42),
    "MultinomialNB":MultinomialNB()
}#here add 5 diffreint model for test the data set 

for model_name,model in models.items():#items for print both key and value in dict
    #print(model_name, ":", model)
    #train model
    model.fit(X_train,y_train)
    # test_model
    predictions = model.predict(X_test)
    #calculate accuracy
    accuracy=accuracy_score(y_test, predictions) 
    #calculate CM
    cm = confusion_matrix(y_test,predictions)
    
    print(f"{model_name} accuracy :{accuracy}")
    print(f"{model_name} confusion matrix:")
    print(np.array2string(cm, separator=', '))
```

```python
#all model give best result 
#i chose svc model 
svc= SVC(kernel="linear")
svc.fit(X_train,Y_train)
ypred=svc.predict(X_test)
accuracy_score(Y_test,ypred)
```

```python
from spellchecker import SpellChecker

# Function to correct spelling of the symptoms
def correct_spelling(user_symptoms):
    spell = SpellChecker()
    corrected_symptoms = []
    
    for symptom in user_symptoms:
        # Find the most likely correction for each symptom
        corrected = spell.correction(symptom)
        corrected_symptoms.append(corrected)
    
    return corrected_symptoms

# Test 1
symptoms = input("Enter your symptoms (comma-separated)...: ")

# Convert the input string to lowercase and split by commas
user_symptoms = symptoms.lower()
user_symptoms = [s.strip() for s in user_symptoms.split(',')]

# Correct the spelling of each symptom
user_symptoms = correct_spelling(user_symptoms)

# Print the corrected symptoms
print("Corrected symptoms:", user_symptoms)

# Proceed with prediction and other logic
predicted_disease = get_predicted_value(user_symptoms)
desc, pre, med, die, workout = helper(predicted_disease)

# Print predicted disease
print("================= Predicted Disease ============")
print(predicted_disease)

# Print predicted description
print("================= Description ==================")
print(desc)

# Print precautions
print("================= Precautions ==================")
i = 1
for p_i in pre[0]:
    print(i, ":", p_i)
    i += 1

# Print medications
print("================= Medications ==================")
i = 1
for m_i in med:
    print(i, ":", m_i)
    i += 1

# Print workout
print("================= Workout ======================")
i = 1
for w_i in workout:
    print(i, ": ", w_i)
    i += 1

# Print diets
print("================= Diets ========================")
i = 1
for d_i in die:
    print(i, ": ", d_i)
    i += 1
```
